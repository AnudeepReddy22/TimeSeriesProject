**Exercise 1**

Generate AR processes with with order = 1 and at least 3 different phi values. Plot the ACF and PACF graphs of these processes and comment on how the graphs change with phi values 

```python
#import the necessary modules 
from statsmodels.tsa.arima_process import ArmaProcess
import numpy as np
ar1 = np.array([1, -0.5])
ma1 = np.array([1])
AR_object1 = ArmaProcess(ar1, ma1)
simulated_data = AR_object1.generate_sample(nsample=600)
#plot the simulated data using matplotlib 
import matplotlib.pyplot as plt 
plt.plot(simulated_data)
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
#plot the ACF and PACF 
plot_acf(simulated_data, lags= 100);
plot_pacf(simulated_data, lags = 50);

# ACF:
# The ACF will show a slow decrease as the lag increases.

# PACF:
# The PACF will have a significant value at lag 1, which means that y(t) is highly correlated with y(t-1).
# The PACF will be close to zero for all other lags, which means that y(t) is not correlated with other lagged values.
# The zero lag is correlated with itself which is to be ignored
```
